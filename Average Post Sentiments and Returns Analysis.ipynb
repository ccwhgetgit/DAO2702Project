```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import data
data = pd.read_csv("Datasets/Combined.csv")
data.head(20)
```

```python
# Transform data: price -> percentage change
percent_change_data = data.copy()
percent_change_data["S&P500 Price"] = percent_change_data["S&P500 Price"].pct_change()
percent_change_data["DJI Price"] = percent_change_data["DJI Price"].pct_change()
percent_change_data["NASDAQ Price"] = percent_change_data["NASDAQ Price"].pct_change()
percent_change_data["USSC2000 Price"] = percent_change_data["USSC2000 Price"].pct_change()
percent_change_data
```

```python
# Post Sentiments vs S&P500

plt.plot(percent_change_data["Date"], percent_change_data["S&P500 Price"], label="S&P500 Price", color="b")
plt.plot(percent_change_data["Date"], percent_change_data["DJI Price"], label="DJI Price", color="g")
plt.plot(percent_change_data["Date"], percent_change_data["NASDAQ Price"], label="NASDAQ Price", color="r")
plt.plot(percent_change_data["Date"], percent_change_data["USSC2000 Price"], label="USSC2000 Price", color="y")

plt.plot(percent_change_data["Date"], percent_change_data["Average"], label="Average Sentiments", color="m")

plt.title("Percentage in Price Change vs Average Sentiments")
plt.xlabel("Time")
plt.xticks("")
plt.legend()
```
