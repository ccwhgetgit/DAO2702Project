```python
import pandas as pd
```

```python
# Read and clean S&P500 Data
SP500_data = pd.read_csv("Datasets/S&P 500 Historical Data.csv")
SP500_data["Date"] = pd.to_datetime(SP500_data["Date"])
SP500_data = SP500_data.drop(columns="Open").drop(columns="High").drop(columns="Low").drop(columns="Vol.").drop(columns="Change %")
SP500_data = SP500_data.rename(columns= {"Price": "S&P500 Price"})
# SP500_data = SP500_data.sort_values(by=["Date"], ascending=True)
SP500_data.head()
```

```python
# Read and clean DJI Data
DJI_data = pd.read_csv("Datasets/Dow Jones Industrial Average Historical Data.csv")
DJI_data["Date"] = pd.to_datetime(DJI_data["Date"])
DJI_data = DJI_data.drop(columns="Open").drop(columns="High").drop(columns="Low").drop(columns="Vol.").drop(columns="Change %")
DJI_data = DJI_data.rename(columns= {"Price": "DJI Price"})
# DJI_data = DJI_data.sort_values(by=["Date"], ascending=True)
DJI_data.head()
```

```python
# Read and clean NASDAQ Data
NASDAQ_data = pd.read_csv("Datasets/NASDAQ Composite Historical Data.csv")
NASDAQ_data["Date"] = pd.to_datetime(NASDAQ_data["Date"])
NASDAQ_data = NASDAQ_data.drop(columns="Open").drop(columns="High").drop(columns="Low").drop(columns="Vol.").drop(columns="Change %")
NASDAQ_data = NASDAQ_data.rename(columns= {"Price": "NASDAQ Price"})
# NASDAQ_data = NASDAQ_data.sort_values(by=["Date"], ascending=True)
NASDAQ_data.head()
```

```python
# Read and clean USSC2000 Data
USSC2000_data = pd.read_csv("Datasets/US Small Cap 2000 Historical Data.csv")
USSC2000_data["Date"] = pd.to_datetime(USSC2000_data["Date"])
USSC2000_data = USSC2000_data.drop(columns="Open").drop(columns="High").drop(columns="Low").drop(columns="Vol.").drop(columns="Change %")
USSC2000_data = USSC2000_data.rename(columns= {"Price": "USSC2000 Price"})
# USSC2000_data = USSC2000_data.sort_values(by=["Date"], ascending=True)
USSC2000_data
```

```python
# Read reddit daily data
reddit_data = pd.read_csv("Datasets/reddit_wsb_daily.csv")
# reddit_data = reddit_data.drop(columns="Unnamed: 0")
reddit_data = reddit_data.rename(columns={"date": "Date", "number": "Number", "avg": "Avg"})
reddit_data["Date"] = pd.to_datetime(reddit_data["Date"])
reddit_data.head()
```

```python
# Merging data from S&P500, DJI, NASDAQ, USSC2000
financial_df = pd.DataFrame({"Date": pd.date_range("2021-01-01", "2021-08-31")})
financial_df = financial_df.merge(SP500_data).merge(DJI_data).merge(NASDAQ_data).merge(USSC2000_data)
financial_df = financial_df.merge(reddit_data)
# financial_df = financial_df.drop(columns=financial_df.columns[0])
financial_df
```

```python
Financial_data = financial_df.to_csv("Datasets/Financial_Data.csv", index=False)
```
