```python
import datetime
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from nltk.sentiment import SentimentIntensityAnalyzer
from random import shuffle


# Import data
data = pd.read_csv('Combined.csv')


#normalisation 
def normalize_data(df):
    minimum = min(df)
    maximum = max(df)
    x = df
    y = (x - minimum) / (maximum  - minimum)
    return y

data
```

```python
#Number of Threads v S&P500 Returns

plt.bar(data['Date'], normalize_data(data['Count']),label = 'Number of Threads', width= 0.6, color = 'r')
plt.plot(data['Date'], normalize_data(data['DJI Price']), label = 'DJI Returns')
plt.xlabel("Dates")
plt.ylabel("Normalized Parameters")
plt.legend()
plt.title("Number of Threads v S&P500 Returns")

```

```python
#Number of Threads v S&P500 Returns

plt.bar(data['Date'], normalize_data(data['Count']),label = 'Number of Threads', width= 0.6, color = 'r')
plt.plot(data['Date'], normalize_data(data['S&P500 Price']), label = 'S&P500 Returns')
plt.plot(data['Date'], normalize_data(data['DJI Price']), label = 'DJI Price', color = 'g')
plt.plot(data['Date'], normalize_data(data['NASDAQ Price']), label = 'NASDAQ Price', color = 'm')
plt.plot(data['Date'], normalize_data(data['USSC2000 Price']), label = 'USSC2000 Price', color = 'y')


plt.xlabel("Dates")
plt.ylabel("Normalized Parameters")
plt.legend()
plt.title("Number of Threads v Various Returns")

data[['Count','DJI Price']].corr()
data[['Count','S&P500 Price']].corr()
data[['Count','USSC2000 Price']].corr()
data[['Count','NASDAQ Price']].corr()
```
